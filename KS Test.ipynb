#K S TEST

KS Test is used to test whether any two distribution are come from the same distribution or not by looking at the p-value.

```python
import numpy as np
import seaborn  as sns
from  scipy import stats
import matplotlib.pyplot as plt
```

```python
#Case 1:
#Generate the r.v  X
x=stats.norm.rvs(size=5000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x),bw=0.5)
plt.show()
```

```python
stats.kstest(x,'norm')     #we can take any distribution in the place of 'norm'
```

Conclusion : As from the KS test results p value is more i.e 47.43% since it is greater than 5 %  , the Ho(Null hypothesis) is true so Generated r.v X and compared normal distribution both are come from same distribution.  

```python
#Case 2:
#Y -continuoes uniform distribution
y=np.random.uniform(0,1,50000)
sns.kdeplot(np.array(y),bw=0.10)
plt.show()
```

```python
stats.kstest(y,'norm')
```

Conclusion : As from the KS test results p value is less i.e 0.0 % since it is less than 5 % , the Ho(Null hypothesis) is not true so Generated r.v Y and compared normal distribution both are not come from same distribution.
